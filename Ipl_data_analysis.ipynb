```python
numeric_columns = df.select_dtypes(include=['number']).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

# For non-numeric columns, fill NaNs with the mode (most frequent value)
non_numeric_columns = df.select_dtypes(exclude=['number']).columns
df[non_numeric_columns] = df[non_numeric_columns].apply(lambda x: x.fillna(x.mode()[0]))
```

```python
df.dropna(inplace=True)
```

**Irregularities in data**

```python
def irregularities(df):
  for i in df:
    print('*'*20,i,'*'*20)
    print(df[i].unique())
    print()
irregularities(df)
```

# Univariate Analysis-Visual

```python
#Univariate Analysis

# Match Outcomes
plt.figure(figsize=(10, 6))
sns.countplot(y='winner', data=df, order=df['winner'].value_counts().index)
plt.title('Match Outcomes')
plt.xlabel('Count')
plt.ylabel('winner')
plt.show()
```

```python
# Distribution of Total Runs
plt.figure(figsize=(10, 6))
sns.histplot(df['target_runs'], bins=20, kde=True)
plt.title('Distribution of Total Runs')
plt.xlabel('target_runs')
plt.ylabel('Frequency')
plt.show()
```

# **MultiVariate Analysis-Visual**

```python
# Multivariate Analysis

# Runs scored by winning team
plt.figure(figsize=(10, 6))
sns.boxplot(x='winner', y='target_runs', data=df)
plt.title('Runs Scored by Winning Team')
plt.xlabel('winner')
plt.ylabel('')
plt.xticks(rotation=90)
plt.show('target_runs')
```

```python
# Correlation Heatmap
plt.figure(figsize=(12, 8))
# Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=['number'])
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# **Code for Toss winning-Match winning Analysis**

```python
# Assuming 'toss_winner' is the column indicating the team that won the toss
df['toss_match_winner'] = df['toss_winner'] == df['winner']

# Display the first few rows to check the new column
print(df[['toss_winner', 'winner', 'toss_match_winner']].head())
```

```python
# Frequency of toss-winning teams that won the match
toss_match_winner_counts = df['toss_match_winner'].value_counts()
print("Toss Winning Team Wins the Match:")
print(toss_match_winner_counts)

# Calculate the percentage
toss_match_winner_percentage = df['toss_match_winner'].mean() * 100
print(f"Percentage of Toss Winning Teams Winning the Match: {toss_match_winner_percentage:.2f}%")
```

**Vizualization for Match winning-Toss winning**

```python
# Visualization for Toss Winning vs. Match Winning
plt.figure(figsize=(10, 6))
sns.barplot(x=toss_match_winner_counts.index, y=toss_match_winner_counts.values, palette='viridis')
plt.title('Toss Winning vs. Match Winning')
plt.xlabel('Toss Winning Team Won the Match')
plt.ylabel('Number of Matches')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])
plt.show()
```

```python
#Count the number of "Man of the Match" awards for each player
man_of_the_match_counts = df['player_of_match'].value_counts()

# Select the top 10 players with the most "Man of the Match" awards
top_players = man_of_the_match_counts.head(10)

# Plot the top players
plt.figure(figsize=(12, 8))
sns.barplot(x=top_players.values, y=top_players.index, palette='viridis')
plt.title('Top 10 Players with Most "Man of the Match" Awards')
plt.xlabel('Number of Awards')
plt.ylabel('Player')
plt.show()
```

# **Vizualization for teams with top 10 score**

```python
# Teams with Top 10 Scores Visualization
# Assuming 'team1' is the team that batted first and 'total_runs' is their score
# Group by match and team to find the highest scores
top_scores = df.groupby(['id', 'team1'])['target_runs'].max().reset_index()

# Sort by the highest scores and take the top 10
top_scores = top_scores.sort_values(by='target_runs', ascending=False).head(10)

# Plot the top scores
plt.figure(figsize=(12, 8))
sns.barplot(x=top_scores['target_runs'], y=top_scores['team1'], palette='viridis')

plt.title('Top 10 Team Scores')
plt.xlabel('Total Runs')
plt.ylabel('Team')
plt.show()
```
